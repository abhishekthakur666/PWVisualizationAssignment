1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic
dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate synthetic dataset
x = np.random.rand(100)
y = 2 * x + np.random.randn(100) * 0.5

# Create a scatter plot using Seaborn
sns.scatterplot(x=x, y=y)
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.title("Scatter Plot of X vs Y")
plt.show()
```

2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate random numbers
data = np.random.normal(loc=0, scale=1, size=1000)

# Create a histogram using Seaborn
sns.histplot(data, kde=True)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Distribution of Random Numbers")
plt.show()
```

3. Create a dataset representing categories and their corresponding values. Compare different categories
based on numerical values.

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Category': ['A', 'A', 'B', 'B', 'C', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
    'Value': [10, 15, 8, 12, 18, 20, 13, 9, 16, 11, 14, 19]
}

# Using Box Plot
sns.boxplot(x='Category', y='Value', data=data)
plt.xlabel("Category")
plt.ylabel("Value")
plt.title("Comparison of Categories based on Value based on box plot")
plt.show()

# Using violin Plot
sns.violinplot(x='Category', y='Value', data=data)
plt.xlabel("Category")
plt.ylabel("Value")
plt.title("Comparison of Categories based on Value based on violin plot")
plt.show()
```

4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical
variable across different categories.

```python

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate a dataset with categories and numerical values
categories = ['A', 'B', 'C', 'D']
data = []
for category in categories:
  for _ in range(50):
    data.append({'Category': category, 'Value': np.random.normal(loc=np.random.randint(10, 30), scale=5)})

df = pd.DataFrame(data)

# Using box plot
sns.boxplot(x='Category', y='Value', data=df)
plt.xlabel("Category")
plt.ylabel("Value")
plt.title("Distribution of Value across Categories (Box Plot)")
plt.show()


# Using violin plot
sns.violinplot(x='Category', y='Value', data=df)
plt.xlabel("Category")
plt.ylabel("Value")
plt.title("Distribution of Value across Categories (Violin Plot)")
plt.show()


# Using Swarm plot
sns.swarmplot(x='Category', y='Value', data=df)
plt.xlabel("Category")
plt.ylabel("Value")
plt.title("Distribution of Value across Categories (Swarm Plot)")
plt.show()
```

5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


# Generate synthetic dataset with correlated features
num_samples = 100
x1 = np.random.rand(num_samples)
x2 = 0.7 * x1 + np.random.randn(num_samples) * 0.1
x3 = -0.5 * x1 + np.random.randn(num_samples) * 0.2
x4 = 0.8 * x2 + np.random.randn(num_samples) * 0.3

# Create a pandas DataFrame
data = pd.DataFrame({'x1': x1, 'x2': x2, 'x3': x3, 'x4': x4})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```
