1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np


output_notebook()

# Create x-values from 0 to 10 with 100 data points
x = np.linspace(0, 10, 100)

# Calculate y-values as the sine of x
y = np.sin(x)

# Create a Bokeh figure
p = figure(title="Sine Wave", x_axis_label="x", y_axis_label="sin(x)")

# Add a line glyph to the figure
p.line(x, y, legend_label="Sine", line_width=2)

# Show the plot
show(p)
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
from bokeh.plotting import figure, show
import numpy as np
import pandas as pd

# Create a DataFrame with random x and y values, sizes, and colors
n = 50
df = pd.DataFrame({
    'x': np.random.rand(n),
    'y': np.random.rand(n),
    'sizes': np.random.randint(5, 20, n),
    'colors': np.random.choice(['red', 'green', 'blue'], n)
})

# Create a Bokeh figure
p = figure(title="Scatter Plot with Size and Color Variations",
           x_axis_label="x", y_axis_label="y")

# Add a scatter glyph to the figure
p.scatter(x='x', y='y', size='sizes', color='colors', source=df)


# Show the plot
show(p)
```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

```python
from bokeh.plotting import figure, show

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a Bokeh figure
p = figure(x_range=fruits, outer_height=350, title="Fruit Counts",
           toolbar_location=None, tools="")

# Add a vbar glyph to the figure
p.vbar(x=fruits, top=counts, width=0.9)

# Customize the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0

# Show the plot
show(p)
```

4. Create a Bokeh histogram to visualize the distribution of the given data.

```python
from bokeh.plotting import figure, show

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title="Histogram", x_axis_label="Value", y_axis_label="Frequency")

# Add a quad glyph to the figure
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="navy", line_color="white", alpha=0.5)

# Show the plot
show(p)
```

5. Create a Bokeh heatmap using the provided dataset.

```python
from bokeh.plotting import figure, show

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create a Bokeh figure
p = figure(title="Heatmap", x_axis_label="x", y_axis_label="y")

# Add a heatmap graph to the figure
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, palette="Viridis256")

# Show the plot
show(p)
```
